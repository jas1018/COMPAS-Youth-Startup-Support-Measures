### geojson 파일 읽기

```python
import folium
import json

map_osm = folium.Map(location=[37.2210864,127.1775537], 
               tiles='cartodbpositron',
               zoom_start=11)

rfile = open('../data/13.용인시_행정경계(읍면동).geojson', 'r', encoding='utf-8').read()
jsonData = json.loads(rfile)
jsonData
```

### csv 파일 처리

```python
# 파일 불러오기

import pandas as pd

df = pd.read_csv("../data/1.용인시_상권_정보.csv")
df
```

```python
df.groupby(by=['대분류코드'], as_index=False).count()
```

```python
# 필요 데이터만 가져오기

test_df = df[["대분류코드", "행정동명"]]
test_df.info()
```

```python
test_df
```

```python
# NaN 값 삭제

test_df = test_df.dropna()
```

```python
# 필터링 

test_df = test_df[test_df["대분류코드"] == 'Q']

test_df = test_df.groupby(by=['행정동명'], as_index=False).count()
```

```python
test_df
```

### 지도에 그리기

```python
folium.Choropleth(
    geo_data = jsonData,
    data = test_df, 
    columns = ('행정동명', '대분류코드'), 
    key_on = 'feature.properties.ADM_DR_NM',
    fill_color = 'YlGn',
    legend_name = '업종별',
).add_to(map_osm)

map_osm
```
